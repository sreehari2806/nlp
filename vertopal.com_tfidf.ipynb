import nltk

import math

nltk.download('stopwords')

from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize

from nltk.stem import PorterStemmer

from collections import Counter

def calculate_tf(word,document):

word_frequency=document.count(word)

return word_frequency/len(document)

def calculate_idf(word,corpus):

num_documents_containing_word=len(\[True for document in corpus if word
in document\])

if num_documents_containing_word==0:

return 0

else:

return math.log10(len(corpus)/num_documents_containing_word)

def calculate_tfidf(document,corpus):

ps=PorterStemmer()

stop_words=set(stopwords.words('english'))

words=\[ps.stem(word.lower()) for word in word_tokenize(document) if
word.lower() not in stop_words\]

word_tfidf_values={}

for word in words:

if word not in word_tfidf_values:

tf=calculate_tf(word,words)

idf=calculate_idf(word,corpus)

word_tfidf_values\[word\]=tf\*idf

return word_tfidf_values

corpus = \[

"this is the first document",

"this document is the second document",

"and this is the third one",

"is this the first document"\]

document="this document is the second document"

tfidf_vector=calculate_tfidf(document,corpus)

print(tfidf_vector)

import nltk  
nltk.download('stopwords')  
nltk.download('words')  
from nltk.corpus import stopwords  
from nltk.tokenize import word_tokenize  
from nltk.stem import PorterStemmer  
from nltk.corpus import words  
text="This is a sample text that we used to demonsstrate NLTK text
processing 123"  
token=word_tokenize(text)  
stop_words=set(stopwords.words("english"))  
stemmer=PorterStemmer()  
alpha_tokens=\[token.lower() for token in token if token.isalpha()\]  
english_words=set(words.words())  
valid_tokens=\[token for token in alpha_tokens if token in
english_words\]  
filtered_tokens=\[token for token in valid_tokens if token not in
stop_words\]  
stemmer_tokens=\[stemmer.stem(token) for token in filtered_tokens\]  
print("Original text:",text)  
print("Tokenisecd text:",token)  
print("Filtered text:",filtered_tokens)  
print("Validated text:",valid_tokens)  
print("Alpha text:",alpha_tokens)  
print("Stemmer text",stemmer_tokens)

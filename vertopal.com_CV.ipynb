**<u>PROGRAM</u>**

import nltk

text=input("Enter the text:")

words=nltk.word_tokenize(text)

num_words=len(words)

num_the=words.count('the')

unique_words=set(words)

num_unique_words=len(unique_words)

percen_unique=(num_unique_words/num_words)\*100

print(words)

print("No: of words:",num_words)

print("No: of occurances of 'the':",num_the)

print("No: of unique words:",num_unique_words)

print("Percentage of unique words:",percen_unique,"%")

**<u>OUTPUT</u>**

Enter the text:Welcome to the world  
\['Welcome', 'to', 'the', 'world'\]  
No: of words: 4  
No: of occurances of 'the': 1  
No: of unique words: 4  
Percentage of unique words: 100.0 %
